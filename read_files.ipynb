```python
import healpy as hp
from astropy.io import fits
```

Download the PR4_NILC_y_map.fits and Masks.fits from https://doi.org/10.5281/zenodo.7940376.

# Compton y-parameter maps

```python
# Read the map files using healpy.

hp.read_map("PR4_NILC_y_map.fits", field=(0,1,2), h=True)
```

```python
# Another option is using astropy fits

pr4_ymap_file = fits.open("PR4_NILC_y_map.fits")
print(pr4_ymap_file.info())
pr4_ymap_file[1].header
```

```python
# Visualize the maps

hp.orthview(pr4_ymap_file[1].data['FULL'], title='NILC y-map', unit='y', min = -8e-6, max = 8e-6, cmap='jet', rot=(0,90,0))
hp.mollview(pr4_ymap_file[1].data['HALF-RING 1'], title='NILC HR1 y-map', unit='y', min = -8e-6, max = 8e-6, cmap='jet')
hp.mollview(pr4_ymap_file[1].data['HALF-RING 2'], title='NILC HR2 y-map', unit='y', min = -8e-6, max = 8e-6, cmap='jet')
```

# Masks

```python
# Read the masks using healpy.

hp.read_map("Masks.fits", field=(0,1,2), h=True)
```

```python
# Read using astropy fits

masks = fits.open("Masks.fits")
masks[1].header
```

```python
hp.mollview(masks[1].data['NILC-MASK'])
hp.mollview(masks[1].data['GAL-MASK'], norm='hist')
hp.mollview(masks[1].data['PS-MASK'])
```

```python
import numpy as np
# Print the sky fraction of the masks

print("NILC mask sky fraction: ", np.mean(masks[1].data['NILC-MASK']**2))
print("GAL mask sky fraction: ", np.mean(masks[1].data['GAL-MASK']**2))
print("PS mask sky fraction: ", np.mean(masks[1].data['PS-MASK']**2))
print("GAL-MASK + PS-MASK sky fraction: ", np.mean((masks[1].data['GAL-MASK'] * masks[1].data['PS-MASK'])**2))
```
